```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands=mp.solutions.hands
```

```python
cap= cv2.VideoCapture(0)
#deschid modulul mediapipe
with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    #in bucla citesc fiecare frame
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        image = cv2.flip(image, 1)
        
        image.flags.writeable = False

        #detectezi aparitia unor maini
        results = hands.process(image)
        
        image.flags.writeable = True
        
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        print(results)
        
        #daca se detecteaza o mana desenez landmark-urile si conexiunile
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                        mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2),
                                         )
            
        
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
 
```

```python
def get_label(index, hand, results):
    output = None
    #parcurg toate etichetele
    for idx, classification in enumerate(results.multi_handedness):
        #daca eticheta se portiveste pentru mana stanga/dreapta
        if classification.classification[0].index == index:
            
            # extrag textul
            label = classification.classification[0].label #stanga sau dreapta
            score = classification.classification[0].score #cat de sigur e mana aia
            text = '{} {}'.format(label, round(score, 2))
            
            # extrag coordonatele incheieturii ca sa stiu unde pun textul
            coords = tuple(np.multiply(
                np.array((hand.landmark[mp_hands.HandLandmark.WRIST].x, hand.landmark[mp_hands.HandLandmark.WRIST].y)),
            [640,480]).astype(int))
            
            output = text, coords
            
    return output
```

```python
def draw_finger_angles(image, results, joint_list):
    
    for hand in results.multi_hand_landmarks:
        for joint in joint_list:
            a = np.array([hand.landmark[joint[0]].x, hand.landmark[joint[0]].y]) 
            b = np.array([hand.landmark[joint[1]].x, hand.landmark[joint[1]].y])
            c = np.array([hand.landmark[joint[2]].x, hand.landmark[joint[2]].y])
            
            radians = np.arctan2(c[1] - b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
            angle = np.abs(radians*180.0/np.pi)
            
            if angle > 180.0:
                angle = 360-angle
                
            cv2.putText(image, str(round(angle, 2)), tuple(np.multiply(b, [640, 480]).astype(int)),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    return image
```

```python
def detect_gesture(results, joint_list):
    if not results.multi_hand_landmarks:
        return None
    
    hand = results.multi_hand_landmarks[0]
    
    fingers_up = []
    
    thumb_tip = hand.landmark[4].x
    thumb_mcp = hand.landmark[2].x
    fingers_up.append(1 if abs(thumb_tip - thumb_mcp) > 0.05 else 0)
    
    tip_ids = [8, 12, 16, 20] 
    for tip_id in tip_ids:
        if hand.landmark[tip_id].y < hand.landmark[tip_id - 2].y:
            fingers_up.append(1)
        else:
            fingers_up.append(0)
    
    if fingers_up == [1, 1, 1, 1, 1]:
        return "Palma deschisa"
    elif fingers_up == [0, 0, 0, 0, 0]:
        return "Pumn"
    elif fingers_up == [0, 1, 1, 0, 0]:
        return "Peace"
    elif fingers_up == [0, 1, 0, 0, 0]:
        return "Pointing"
    elif fingers_up == [1, 0, 0, 0, 1]:
        return "Rock on!"
    
    return f"Custom: {sum(fingers_up)} degete"
```

```python
import mediapipe as mp
import cv2
import numpy as np
from collections import deque

mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

class GestureDetector:
    def __init__(self):
        self.gesture_history = deque(maxlen=10)
    
    def detect_fingers_up(self, hand):
        
        fingers = []
        
        thumb_tip = hand.landmark[4]
        thumb_mcp = hand.landmark[2]
        fingers.append(1 if abs(thumb_tip.x - thumb_mcp.x) > 0.05 else 0)
        
        tip_ids = [8, 12, 16, 20]
        for tip_id in tip_ids:
            if hand.landmark[tip_id].y < hand.landmark[tip_id - 2].y:
                fingers.append(1)
            else:
                fingers.append(0)
        
        return fingers
    
    def recognize_gesture(self, hand):
        
        fingers = self.detect_fingers_up(hand)
        
        gestures = {
            (1, 1, 1, 1, 1): ("Palma", (0, 255, 0)),
            (0, 0, 0, 0, 0): ("Pumn", (0, 0, 255)),
            (0, 1, 1, 0, 0): ("Peace", (255, 255, 0)),
            (0, 1, 0, 0, 0): ("Pointing", (255, 0, 255)),
            (1, 0, 0, 0, 1): ("Rock", (255, 100, 0)),
            (1, 1, 0, 0, 0): ("OK", (0, 255, 255)),
        }
        
        gesture_tuple = tuple(fingers)
        if gesture_tuple in gestures:
            return gestures[gesture_tuple]
        
        return (f"{sum(fingers)} degete", (200, 200, 200))
    
    def calculate_pinch_distance(self, hand):
        thumb_tip = hand.landmark[4]
        index_tip = hand.landmark[8]
        
        distance = np.sqrt(
            (thumb_tip.x - index_tip.x)**2 + 
            (thumb_tip.y - index_tip.y)**2
        )
        return distance

def draw_enhanced_ui(image, gesture_text, gesture_color, fps=0):
    h, w, _ = image.shape
    
    overlay = image.copy()
    cv2.rectangle(overlay, (10, 10), (w-10, 120), (0, 0, 0), -1)
    cv2.addWeighted(overlay, 0.3, image, 0.7, 0, image)
    
    cv2.putText(image, "HAND GESTURE TRACKER", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    
    cv2.putText(image, f"FPS: {fps}", (w-120, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
    
    if gesture_text:
        cv2.putText(image, f"Gest: {gesture_text}", (20, 80),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, gesture_color, 2)
    
    cv2.putText(image, "Q - Quit | S - Screenshot", (20, 110),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (200, 200, 200), 1)
    
    return image

def calculate_finger_angle(a, b, c):
    
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180.0:
        angle = 360 - angle
    return angle

def main():
    cap = cv2.VideoCapture(0)
    detector = GestureDetector()
    
    prev_time = 0
    
    joint_list = [[8,7,6], [12,11,10], [16,15,14], [20,19,18]]
    
    with mp_hands.Hands(
        min_detection_confidence=0.7,
        min_tracking_confidence=0.7,
        max_num_hands=2
    ) as hands:
        
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            
            curr_time = cv2.getTickCount()
            fps = int(cv2.getTickFrequency() / (curr_time - prev_time)) if prev_time > 0 else 0
            prev_time = curr_time
            
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image = cv2.flip(image, 1)
            image.flags.writeable = False
            results = hands.process(image)
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            
            gesture_text = None
            gesture_color = (255, 255, 255)
            
            if results.multi_hand_landmarks:
                for num, hand in enumerate(results.multi_hand_landmarks):
                    mp_drawing.draw_landmarks(
                        image, hand, mp_hands.HAND_CONNECTIONS,
                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                        mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2)
                    )
                    
                    gesture_text, gesture_color = detector.recognize_gesture(hand)
                    
                    for joint in joint_list:
                        a = np.array([hand.landmark[joint[0]].x, hand.landmark[joint[0]].y])
                        b = np.array([hand.landmark[joint[1]].x, hand.landmark[joint[1]].y])
                        c = np.array([hand.landmark[joint[2]].x, hand.landmark[joint[2]].y])
                        
                        angle = calculate_finger_angle(a, b, c)
                        
                        coord = tuple(np.multiply(b, [image.shape[1], image.shape[0]]).astype(int))
                        cv2.putText(image, str(int(angle)), coord,
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
                    
                    pinch_dist = detector.calculate_pinch_distance(hand)
                    if pinch_dist < 0.05:  
                        cv2.putText(image, "PINCH!", (20, 150),
                                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            
            image = draw_enhanced_ui(image, gesture_text, gesture_color, fps)
            
            cv2.imshow('Enhanced Hand Tracking', image)
            
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q'):
                break
            elif key == ord('s'):
                cv2.imwrite(os.path.join('Output Images', '{}.jpg'.format(uuid.uuid1())), image)
                print("Screenshot salvat!")
    
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
